```python
import numpy as np
from numpy import nan as NA
import pandas as pd

df = pd.DataFrame(np.random.rand(10, 4))

df.iloc[1,0] = NA
df.iloc[2:3,2] = NA
df.iloc[5:,3]= NA
```

```python
df
```

```python
#리스트와이즈(listwise) 삭제
#NaN이 있는 행을 삭제

df.dropna()
```

```python
#페어와이즈(pairwise) 삭제
#필요한 열을 추출하고 NaN이 있는 행을 삭제

df[[0,1]].dropna()
```

```python
#fillna로 채우기
#fillna(값)로 NaN을 괄호 안의 값으로 대체

df.fillna(0)
```

```python
#이웃하는 값으로 대체하기
#ffill 메서드를 이용해 직전 행의 값으로 대체

df.fillna(method= 'ffill')
```

```python
#평균대체법

#각 컬럼의 평균값

df.mean()
```

```python
df.fillna(df.mean())
```
